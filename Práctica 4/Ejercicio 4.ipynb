```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def calculaMask(img,NDmin,NDmax):
    mask=np.zeros((img.shape[0],img.shape[1]), dtype=np.int16)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if NDmin<img[i][j] and NDmax>img[i][j]:
                mask[i][j]=1
            else:
                mask[i][j]=0
    return mask
    
```

```python
def mascara (img,mask):
    resimg=img.copy()
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            resimg[i][j]=mask[i][j]*img[i][j]
            
    return resimg
```

```python
img = cv2.imread('imgpack/Filomena.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img,cmap='gray') ##Imagen sin tratar
```

```python
mask_luces=calculaMask(img,200,255)
mask_img_luces=mascara(img,mask_luces)
```

```python
plt.imshow(mask_img_luces,cmap='gray') ##Imagen sin tratar
```

```python
mask_nubes=calculaMask(img,100,180)
mask_img_nubes=mascara(img,mask_nubes)
```

```python
plt.imshow(mask_img_nubes,cmap='gray')
```
