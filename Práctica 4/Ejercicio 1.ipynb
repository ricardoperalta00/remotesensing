```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def diezmado(img,R):
    res=img.copy()
    r_new = int(1/R)
    res = img[0:-1:r_new,0:-1:r_new]
    return res
```

```python
img = cv2.imread('imgpack/RAW/6_ISS_Spain_IslaCristina.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img,cmap='gray') ##Imagen sin tratar
```

```python
res001=diezmado(img,0.01)
res005=diezmado(img,0.05)
res01=diezmado(img,0.1)
res025=diezmado(img,0.25)
```

```python
plt.imshow(res001,cmap='gray') 
```

```python
plt.imshow(res005,cmap='gray') 
```

```python
plt.imshow(res01,cmap='gray') 
```

```python
##Ahora las imgenes para guardar con 0,25;0,33 y 0,5 (0,25 ya realizada)
res033=diezmado(img,0.33)
res05=diezmado(img,0.5)
```

```python
cv2.imwrite('ej1_025.png',res025)
cv2.imwrite('ej1_033.png',res033)
cv2.imwrite('ej1_05.png',res05)
```
