```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=np.array([[12,17,32],
              [26,17,40],
              [14,20,27]])
```

```python
def amplia (img):
    resimg=np.zeros((img.shape[0]*2,img.shape[1]*2))
    #TRAZA 1#
    resimg[0:-1:2,0:-1:2]=img #"haciendo uso de slices introducimos los datos"
    #'TRAZA 2'
    for i in range(1,resimg.shape[0]-1,2):
        for j in range(0,resimg.shape[1],2):
            resimg[i,j]=round((resimg[i-1][j]+resimg[i+1][j])/2) #'Recorremos la matriz con índices'
    #'TRAZA 3'
    for i in range(0,resimg.shape[0]):
        for j in range(1,resimg.shape[1]-1,2):
            resimg[i][j]=round((resimg[i][j-1]+resimg[i][j+1])/2) #'Igual que la anterior pero cambiando índices'
    #'TRAZA 4'
    resimg[0:-1,-1]=resimg[0:-1,-2] #'Con slice podemos copiar directamente'
    #'TRAZA 5'
    resimg[-1]=resimg[-2] #'Hacemos que la última fila (-1) se igual que la antepenúltima (-2)'
    return resimg
```

```python
#Comprobamos con un ejemplo pequeño
res = amplia(img)
```
