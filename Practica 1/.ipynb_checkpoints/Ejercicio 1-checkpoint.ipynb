```python
print(numpy.sin(numpy.pi /4) )
x=numpy.arange(0, numpy.pi, 0.01)
print(x)

y= numpy.sin(x)
print(y[0], "...", y[314])

z=numpy.cos(x)
print(z[0], "...", z[314])
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.figure(1)
plt.plot(x)
plt.ylabel('Lista de valores x')
plt.show()
```
