```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('britney-calva.jpg')
print(img)
```

```python
"""
COMENTAR
Es una matrix y los valores en las filas y columnas son los valores de los pixeles

"""
```

```python
plt.imshow(img)
plt.show()
```

```python
print("La imágen es una amtriz de tamaño ",img.shape, "(px, px, canales))")
print("Esto es el pixel [100,100]: ",img[100,100])
print("Esto es la fila 100:", img[100, :])
print("Esto es una submatriz:", img[100:200, 100:200] )
```

```python
b,g,r = cv2.split(img)
img_rgb = cv2.merge ((r,g,b))
```

```python
plt.imshow(img_rgb)
plt.show()
```
