```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def nvdi (path_nir,path_r):
    img_nir = cv2.imread(path_nir, cv2.IMREAD_GRAYSCALE)
    img_r = cv2.imread(path_r, cv2.IMREAD_GRAYSCALE)
    ndvi_res=np.zeros((img_r.shape[0],img_r.shape[1]),np.uint8)
    for i in range(img_r.shape[0]):
        for j in range(img_r.shape[1]):
            ndr=img_r[i][j]
            ndnir=img_nir[i][j]
            if (ndnir + ndr) == 0:
                ndvi_res[i][j]=0
            else:
                ndvi_valor=((ndnir)-(ndr))/((ndnir)+(ndr))
                ndvi_res[i][j] = ((ndvi_valor+1)*127.5).astype(np.uint8) #para escalarlo
                
    return ndvi_res
```

```python
def densi (img, rangos):
    colores={0:[255,0,0],1:[255,140,0],2:[241, 196, 15],3:[46, 204, 113 ],4:[19, 141, 117],5:[52, 152, 219],6:[36, 113, 163],
            7:[125, 60, 152],8:[136, 78, 160],9:[81, 46, 95]}
    res=cv2.merge((img,img,img))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            nd = img[i][j]
            cont = 0
            for r in rangos:
                m = range(r[0],r[1])
                if (nd in m):
                    res[i][j] = colores[cont]
                cont=cont+1
    return res
                

        
        
    
```

```python
rangos = [(128,136),(137,144),(145,152),(153,160),(161,168),(168,176),(177,184),(185,192)]
```

```python
res1 = nvdi ('imgpack/LandSat_C/Landsat_C_05-11_NIR.png','imgpack/LandSat_C/Landsat_C_04-11_R.png')
```

```python
resdens = densi(res1,rangos)
```
