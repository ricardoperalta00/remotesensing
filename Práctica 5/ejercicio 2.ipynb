```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
img = cv2.imread('imgpack/RAW/6_ISS_Spain_IslaCristina.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def seudo(img,_min,_max):
    res=np.zeros((img.shape[0],img.shape[1],3), np.uint8)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            nd = img[i][j]
            if nd>=_min and nd<=_max:
                res[i][j] = [255,0,0]
            else:
                res[i][j] = [int(nd),int(nd),int(nd)]
    return res
    
```

```python
img_res=seudo(img,180,255)
```

```python
plt.imshow(img_res)
```
