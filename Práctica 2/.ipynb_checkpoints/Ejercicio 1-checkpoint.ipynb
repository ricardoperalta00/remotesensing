```python
import cv2
import matplotlib.pyplot as plt
import sys
import numpy as np
```

```python
img = cv2.imread('imgpack/RAW/1_Chicago.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img[0]
```

```python
def Histograma(img):
    " Para crear una lista ordenada con los valores"
    l=[0 for x in range(256)]
    for m in img:
        l2=m.tolist()
        for x in range(256):
            l[x]=l[x]+l2.count(x)
            
    "Para ver el histograma"
    fig, ax = plt.subplots()
    ax.bar(range(len(l)), l)
    plt.show()
    
    
```

```python
Histograma(img)
```
