```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('imgpack/RAW/6_ISS_Spain_IslaCristina.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def filtro(img,cf):
    CFsum=sum(sum(cf))
    resimg=img.copy()
    ##Para poner los bordes a 0
    resimg[0]=0
    resimg[-1]=0
    resimg[:,0]=0
    resimg[:,-1]=0 
    
    ## Este bucle es para ir mirando las submatrices, hago cf.shape[0]-1 para que podamos usar un filtro diferente a 3x3
    ## si no colocaría -2 directamente que es lo mínimo para que no se salga del index
    for i in range(img.shape[0]-cf.shape[0]-1):
        for j in range(img.shape[1]-cf.shape[1]-1):
            ##Con slice cogemos la submatriz la multiplicamos x el filtro y sumamos todo
            ##finalmente sustituimos en resimg
            val=round(sum(sum(img[i:i+3,j:j+3]*cf))/CFsum)
            resimg[i+1,j+1]=val
    return resimg
    
```

```python
plt.imshow(img,cmap='gray') ##Imagen sin tratar
```

```python
cf=np.array([
    [1,1,1],
    [1,1,1],
    [1,1,1]
])
cf2=np.array([
    [1,1,1],
    [1,2,1],
    [1,1,1]
])
cf3=np.array([
    [1,2,1],
    [2,4,2],
    [1,2,1]
])
```

```python
imgmed=filtro(img,cf)
```

```python
imgmed_img=filtro(img,cf2)
```

```python
imggauss=filtro(img,cf3)
```

```python
fig, axs = plt.subplots(2,2)
axs[0,0].imshow(img, cmap='gray')
axs[0,1].imshow(imgmed, cmap='gray')
axs[1,0].imshow(imgmed_img, cmap='gray')
axs[1,1].imshow(imggauss, cmap='gray')
```
