```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=np.array([
            [12,14,17,24,32,30],
            [19,18,17,26,36,40],
            [26,21,17,28,40,42],
            [20,19,18,26,33,35],
            [14,17,20,23,27,20],
            [35,14,22,13,40,21]
    ])
```

```python
def mediana(img):
    resimg=img.copy()
    ##Para poner los bordes a 0
    resimg[0]=0
    resimg[-1]=0
    resimg[:,0]=0
    resimg[:,-1]=0 
    
    ## Este bucle es para ir mirando las submatrices de 3x3
    for i in range(img.shape[0]-2):
        for j in range(img.shape[1]-2):
            val=np.median(img[i:i+3,j:j+3])
            resimg[i+1,j+1]=val
    return resimg
    
```

```python
##Prueba con una matriz peque√±a y vemos que sale todo correcto
mediana(img)
```

```python
img = cv2.imread('imgpack/WorldView_1-1_PAN.png', cv2.IMREAD_GRAYSCALE)
```

```python
resimg = mediana(img)
```

```python
fig, axs = plt.subplots(1,2)
axs[0].imshow(img, cmap='gray')
axs[1].imshow(resimg, cmap='gray')
```

```python
cv2.imwrite('ej3.png',resimg)
```
