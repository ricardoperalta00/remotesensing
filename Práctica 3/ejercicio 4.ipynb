```python
import numpy as np
import math
import cv2
import matplotlib.pyplot as plt
```

```python
img=np.array([
            [12,14,17,24,32,30],
            [19,18,17,26,36,40],
            [26,21,17,28,40,42],
            [20,19,18,26,33,35],
            [14,17,20,23,27,20],
            [35,14,22,13,40,21]
    ])
```

```python
def sobel(img):
    resimg=img.copy()
    resimg[0]=0
    resimg[-1]=0
    resimg[:,0]=0
    resimg[:,-1]=0
    for i in range(img.shape[0]-2):
        for j in range(img.shape[1]-2):
            submat=img[i:i+3,j:j+3].copy()
            C=(sum(submat[0]*(1,2,1)))-sum((submat[2]*(1,2,1)))
            F=(sum(submat[0:3,2]*(1,2,1)))-(sum(submat[0:3,0]*(1,2,1)))
            resimg[i+1,j+1]=round(math.sqrt(C**2+F**2))
    return resimg
```

```python
sobel(img)
```

```python
img = cv2.imread('imgpack/RAW/6_ISS_Spain_IslaCristina.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
resimg = sobel(img)
```

```python
plt.imshow(img,cmap='gray') ##Imagen sin tratar
```
